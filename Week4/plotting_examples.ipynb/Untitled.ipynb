```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.title('Gas Prices over Time (in USD)')
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight': 'bold', 'fontsize': 25})
plt.plot(gas.Year, gas.USA, 'b.-', label='United States') 
plt.plot(gas.Year, gas.Canada, 'y.-', label= 'Canada')
plt.plot(gas.Year, gas['South Korea'], 'r.-', label= 'South Korea')
plt.plot(gas.Year, gas.Australia, 'g.-', label='Australia')



#Different ways to plot values

# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('gas_price_figure.png', dpi=300)

plt.show()
```

```python
### Load Fifa Data
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.ylabel('Distribution of Player Skills in FIFA 2018')

plt.savefig('hist_fifa_players.png', dpi=300)

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['left', 'right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f')
plt.title('Foot Preference of FIFA Players')

plt.savefig('pie_chart_fifa_foot.png', dpi=300)

plt.show()
```

```python

fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
medium_light = fifa.loc[(fifa.Weight > 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >=150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, medium_light, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.1,.1,.1,.4)

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.savefig('pie_chart_weight.png', dpi=300)

plt.show()
```

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']


plt.boxplot([barcelona, madrid, revs], labels=labels)

plt.title('FIFA Team Comparisons')
plt.ylabel('FIFA Overall Rating')

plt.savefig('boxplot_team.png', dpi=300)
plt.show()
```
